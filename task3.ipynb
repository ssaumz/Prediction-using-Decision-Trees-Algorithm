```python
X = df.iloc[:, [0,1,2,3]].values

Y = iris.target

print(X)

Y
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier(random_state = 100)
dtree.fit(df,y)


print('Decision Tree classifier Created')

dtree
```

```python
#Spliting our data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=100)
```

```python
dtree.fit(X_train, y_train)
dtree
```

```python
#Now to predict on random data

X = [[5.4, 3.9, 1.3, 0.4]]
Y_pred = dtree.predict(X)

print(Y_pred)
```

```python
#Prediction on X_test (Testing data)

Y_pred = dtree.predict(X_test)
print(Y_pred)
```

```python
#Checking Accuracy of our model

import sklearn.metrics as metrics

print("Accuracy:",metrics.accuracy_score(y_test, Y_pred))

#Checking with the help of Confusion Matrix

cfm = np.array(metrics.confusion_matrix(y_test, Y_pred))

print(cfm)
```

```python
# Import necessary libraries
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

# Load dataset
iris = load_iris()
X, y = iris.data, iris.target

# Train a DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree.fit(X, y)

# Plot the tree using matplotlib
plt.figure(figsize=(20, 10))
tree.plot_tree(dtree, feature_names=iris.feature_names, class_names=iris.target_names, filled=True)
plt.show()
```

```python
#We can also see the decisons made in the Decision Tree

from sklearn import tree

txt_rep = tree.export_text(dtree)

print(txt_rep)
```
